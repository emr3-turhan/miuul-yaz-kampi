> **24-Uygulama Mülakat Sorusu**

```python
soru = "hi my name is john and i am learning python"
```

```python
def alternating(string):
    new_soru = ""
    for i in range(len(string)):
        if (i % 2) == 0:
            new_soru += string[i].upper()
        else:
            new_soru += string[i].lower()
    print(new_soru)
alternating(soru)
```

> **26-Enumerate**

```python
students = ["John","Mark","Venessa","Mariam"]
```

```python
for student in students:
    print(student)
```

```python
A=[]
B=[]

for index, student in enumerate(students):
    if index % 2 == 0:
        A.append(student)
    else:
        B.append(student)
    
```

```python
A
```

```python
B
```

```python
def divide_students(students):
    groups = [[],[]]
    for index,student in enumerate(students):
        if index % 2 == 0:
            groups[0].append(student)
        else:
            groups[1].append(student)
    
    
    
    return groups
    
divide_students(students)
```

```python
values = [10,30]
result = lambda values: [i for i in range(min(values),max(values)) if True in [False for j in range(2,i) if i % j == 0]]
```

```python
result(values)
```

```python
res = lambda values: [x for x in range(min(values),max(values)) if all(x%y != 0 for y in range(2,x))]
```
