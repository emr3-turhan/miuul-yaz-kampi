```python
# Datetime yapalım
dff['Timestamp'] = pd.to_datetime(dff['Timestamp'],format='%Y-%m-%d')
```

```python
dff.info()
```

```python
# Timestamp değişkeni görüldüğü üzere datetime dtype'ında oldu.
# Artık yıl,ay,gün gibi özellikleri alabiliriz.

# Yıl
dff['year'] = dff['Timestamp'].dt.year

# Ay
dff['month'] = dff['Timestamp'].dt.month

# Gün
dff['day'] = dff['Timestamp'].dt.day

# Yıl farkı
dff["year_fark"] = date.today().year - dff["Timestamp"].dt.year

# Ay farkı
dff["month_fark"] = (date.today().year - dff["Timestamp"].dt.year) * 12 + date.today().month - dff["Timestamp"].dt.month

# Gün adı
dff["day_name"] = dff["Timestamp"].dt.day_name()
```
